```
# get data files
!wget https://cdn.freecodecamp.org/project-data/sms/train-data.tsv
!wget https://cdn.freecodecamp.org/project-data/sms/valid-data.tsv

train_file_path = "train-data.tsv"
test_file_path = "valid-data.tsv"
```

```
train_data = pd.read_csv(train_file_path, sep='\t', header=None, names=['label', 'message'])
test_data = pd.read_csv(test_file_path, sep='\t', header=None, names=['label', 'message'])


from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder


vectorizer = CountVectorizer()

X_train = vectorizer.fit_transform(train_data['message'])
# Transform the test messages
X_test = vectorizer.transform(test_data['message'])
```

```
# Initialize LabelEncoder
label_encoder = LabelEncoder()
# Fit and transform the training labels
y_train = label_encoder.fit_transform(train_data['label'])
# Transform the test labels
y_test = label_encoder.transform(test_data['label'])

# Define and train the model
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)
```

```
# function to predict messages based on model
# (should return list containing prediction and label, ex. [0.008318834938108921, 'ham'])
def predict_message(pred_text):
    pred_text_transformed = vectorizer.transform([pred_text])

    probabilities = model.predict_proba(pred_text_transformed)[0]
    predicted_class = model.predict(pred_text_transformed)[0]
    predicted_label = label_encoder.inverse_transform([predicted_class])[0]

    return [probabilities[1], predicted_label]

pred_text = "how are you doing today?"

prediction = predict_message(pred_text)
print(prediction)
```
